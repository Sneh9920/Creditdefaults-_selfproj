```python
x=df.iloc[:,6:20]
y=df['default payment next month']
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE()
x_smote,y_smote=smote.fit_resample(x,y)
```

```python
y_smote.value_counts()
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_smote,y_smote,test_size=0.2,random_state=0)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)
print(metrics.classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
```

Decision Tree

```python
DTC = DecisionTreeClassifier()
DTC.fit(x_train, y_train)
y_pred2 = DTC.predict(x_test)
print(metrics.classification_report(y_test, y_pred2))
```

Randomforest

```python
randomForest= RandomForestClassifier(n_estimators=1000)
randomForest.fit(x_train,y_train)
y_pred3 = randomForest.predict(x_test)
print(metrics.classification_report(y_test, y_pred3))
```

```python
feat_cols = ['ID', 'LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE', 'PAY_0',
       'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6', 'BILL_AMT1', 'BILL_AMT2',
       'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1',
       'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']
```

```python
classifier = tf.compat.v1.estimator.DNNClassifier(hidden_units = [10, 20, 10], n_classes = 2, feature_columns = feat_cols )
```

```python
input_func = tf.compat.v1.estimator.inputs.pandas_input_fn(x=x_train,y=y_train,batch_size=20,shuffle=True)
```
